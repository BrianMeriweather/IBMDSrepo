### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
print(df.columns)
Pay_Mass= ("SELECT sum (PAYLOAD_MASS__KG_) from SPACEXTBL WHERE Customer = 'NASA (CRS)'")
cur.execute(Pay_Mass)
print (cur.fetchall())
```

### Task 4

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000

### Task 7

##### List the total number of successful and failure mission outcomes


```python
Successful = ("SELECT COUNT (*) FROM SPACEXTBL WHERE Mission_Outcome = 'Success'")
Unsuccessful = ("SELECT COUNT (*) FROM SPACEXTBL WHERE Mission_Outcome != 'Success'")
cur.execute(Successful)
print (cur.fetchall())
cur.execute(Unsuccessful)
print (cur.fetchall())
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery

### Task 9

##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 4, 2) as month to get the months and substr(Date,7,4)='2015' for year.**


```python

Boo1 = ("SELECT substr (Date,4,2) AS month, substr (Date,7,4) as year,Date,Landing_Outcome, Booster_Version, Launch_Site FROM SPACEXTBL WHERE Landing_Outcome like 'Failure (drone ship)' AND year IN (Select substr (Date,7,4) as year FROM SPACEXTBL WHERE year = '2015')")

print ("----------")
cur.execute(Boo1)
Failure_dates= (cur.fetchall())
print (Failure_dates)
```

```python
Monthnamelink = ("https://www.studytonight.com/python-howtos/how-to-get-month-name-from-month-number-in-python")
month_dic={0: 'Jan', 1: 'Feb', 2: 'Mar', 3: 'Apr', 4: 'May', 5: 'Jun',
          6: 'Jul', 7: 'Aug', 8: 'Sep', 9: 'Oct', 10: 'Nov', 11: 'Dec'}
import datetime

Failure_date_list =[]

for date in Failure_dates:
    month_num=date[0]
    datetime_object=datetime.datetime.strptime(month_num,"%m")
    month_name=datetime_object.strftime("%B")
    y=[month_name,date[1],date[3],date[4],date[5]]
    Failure_date_list.append (y)

print(Failure_date_list)
```

### Task 10

##### Rank the  count of  successful landing_outcomes between the date 04-06-2010 and 20-03-2017 in descending order.


```python
Rank=("SELECT Date, Landing_Outcome FROM SPACEXTBL WHERE Landing_Outcome like 'Success' AND Date between '04-06-2010' and '20-03-2017' GROUP BY Landing_Outcome ORDER By Date desc")

cur.execute (Rank)
print (cur.fetchall())
```

```python
Rank=("SELECT Landing_Outcome, COUNT (Landing_Outcome), Date FROM SPACEXTBL WHERE  Date between '04-06-2010' and '20-03-2017' GROUP BY Landing_Outcome ORDER By COUNT(Landing_Outcome) desc")

cur.execute (Rank)
print (cur.fetchall())
```
